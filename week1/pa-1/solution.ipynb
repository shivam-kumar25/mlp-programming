# MetaData

 `Destination`: The destination where the service ends 

 `Route`: Route the flight took. 

 `Dep_Time`: The time when the journey starts from the source. 

 `Arrival_Time`: Time of arrival at the destination. 

 `Duration`: Total duration of the flight. 

 `Total_Stops`: Total stops between the source and destination. 

 `Additional_Info`: Additional information about the flight 

 `Price`: The price of the ticket ▶▶▶ Target

 `WeekDay`: Day at which journey started.

Q1. What is the average of the flight ticket price? Write your answer correct to two decimal places. 


```python
total_sum = df["Price"].sum()
print("Total sum of prices:", total_sum)

print(df.shape)
total_number_of_rows = df.shape[0]
print("Total number of rows:", total_number_of_rows)

average_price = df["Price"].mean()
print("Average price:", average_price)    
print("Answer to the question: The average price of the products in the dataset is : ", average_price.round(2))      
```

Q2: During which month did the highest number of flights occur? Months are represented by numerical codes, with January corresponding to 1, February to 2, and so forth.

- January
- May
- October
- December



```python
# get the month series
month_series = df["Month"]
print("Month series:")
display.display(month_series)

month_series.mode()
# return the most frequent month in the dataset
most_frequent_month = month_series.mode()[0]
print("Most frequent month:", most_frequent_month)
```

Q3: Is the average price of flight tickets higher on weekends (Saturday and Sunday) or on weekdays (Remaining 5 days)?

- Weekends (Saturday and Sunday)
- Weekdays

```python
weekday_and_prices_data = df[["WeekDay", "Price"]]
# display the WeekDay and Price columns
print("Weekday and Prices data:")
display.display(weekday_and_prices_data)       



weekdays_price_ = weekday_and_prices_data.groupby("WeekDay")["Price"]
display.display(weekdays_price_)
weekdays_price_mean = weekdays_price_.mean()
print("Average price for each weekday:")
display.display(weekdays_price_mean)   


   
weekend_price_mean = weekdays_price_mean[["Saturday", "Sunday"]].mean()
print("Average price on weekends:", weekend_price_mean) 



weekdays_price_mean[["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"]].mean()
print("Average price on weekdays:", weekdays_price_mean[["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"]].mean())     



```
