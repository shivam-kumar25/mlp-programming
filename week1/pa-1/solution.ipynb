Q5: Convert the values of 'Duration' into seconds. Enter the average duration (in seconds) of a flight. Enter your answer correct to two decimal places. 



```python

duration_series = df["Duration"]
print("Duration series:")   
display.display(duration_series) 

# format 26h 35m
def convert_to_second(duration : str ) :
       # strt with 0h and 0m
       hours = 0 
       minutes = 0 
       
       if "h" in duration:
              hours = int(duration.split("h")[0])
              
       if "m" in duration:
              minutes = int(duration.split("m")[0].split("h")[-1])    

       
       return str(hours*3600 + minutes * 60) 




test_duration = convert_to_second("26h 35m")
print(test_duration)

duration_series_in_seconds = duration_series.apply(convert_to_second)
display.display(duration_series_in_seconds)

average_duration_in_seconds = duration_series_in_seconds.astype(int).mean()
print("Average duration in seconds:", average_duration_in_seconds)    
```

Apply the following functions to the columns Dep_Time and Arrival_Time:

Transform the values in the 'dep_time' and 'arrival_time' columns to represent the hour component. For instance, if an entry is 10:05 June 13 or 10:05, the corresponding value should be 10. 

Then convert the time into four categories as follows:

5 <= hour < 12 = Morning 

12 <= hour < 17 = Afternoon 

17 <= hour < 20 = Evening 

20 <= hour < 5 = Night

Note: Please ensure that you make the changes directly within the dataset and continue to use that modified dataset for subsequent questions.

Q6: How many flights started in the Morning and arrived the destination at Evening? 


```python
import datetime as dt

df_hour = df[["Dep_Time" , "Arrival_Time"]]

df_hour["Arrival_Time"] = df_hour["Arrival_Time"].apply( lambda time : time.strip().split(" ")[0])

# display.display(df_hour)


# Convert to time fomrat 

df_hour["Arrival_Time"] = pd.to_datetime(df_hour["Arrival_Time"] , format="%H:%M").dt.hour
df_hour["Dep_Time"] = pd.to_datetime(df_hour["Dep_Time"] , format="%H:%M" ).dt.hour
# display.display(df_hour)


# a function to categorise 
def categorise(hour):
       if 5 <= hour < 12 :
              return "Morning"
       
       if 12 <= hour < 17 :
              return "Afternoon"
              

       if 17 <= hour < 20 :
              return  "Evening" 

       if 20 <= hour or hour < 5 :
              return "Night"
       
       
# now apply the function to both the row 
df_hour["Arrival_Time_name"] = df_hour["Arrival_Time"].apply(categorise)
df_hour["Dep_Time_name"] = df_hour["Dep_Time"].apply(categorise)
 
# display.display(df_hour)


# now count 
df_required = df_hour[ (df_hour["Arrival_Time_name"] == "Evening")   & (df_hour["Dep_Time_name"] == "Morning")]

display.display(df_required)
```

Q7: Encode the values of column 'WeekDay' as follows:
Weekends (Sunday, Saturday) = 1
all remaining five days = 0
What is the most frequent (mode) WeekDay?

- Weekends
- Weekdays

```python
df_weekday = df["WeekDay"]

display.display(df_weekday)

def encode_0_1_ (weekday):
       if weekday == "Sunday" or weekday == "Saturday" :
              return 1 
       
       return 0


df_weekday_encoded = df_weekday.apply(encode_0_1_)
display.display(df_weekday_encoded)

print(df_weekday_encoded.mode())
```
