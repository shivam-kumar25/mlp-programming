# Line Plots
Line plots are used to display data points connected by straight lines. They are useful for showing trends over time or continuous data.





### plt.plot() 

plot() is a function in the Matplotlib library that is used to create line plots. It takes in data points and connects them with straight lines to visualize the relationship between the variables. The basic syntax for plt.plot() is as follows:

```python
plt.plot(x, y, format_string, **kwargs)
```    
- x: The data points for the x-axis.
- y: The data points for the y-axis.      
- format_string: An optional string that specifies the format of the plot (e.g., color, marker style, line style).
- **kwargs: Additional keyword arguments that can be used to customize the plot (e.g., labels, title, grid).    



````python
from matplotlib import pyplot as plt

# sample_data_points : (2,1),(5,6),(10,8)
X=[2, 5, 10]
Y=[1, 6, 8]

#create the Plot
plt.plot(X,Y,'ko--')

#Display the Plot
plt.show()
````

# Scatter Plots 
Scatter plots are used to display the relationship between two variables by plotting individual data points on a Cartesian plane. Each point represents a pair of values from the two variables, allowing you to visually assess patterns, correlations, and trends in the data. Scatter plots are particularly useful for identifying clusters, outliers, and the overall distribution of data points. 



### plt.scatter()    

scatter() is a function in the Matplotlib library that is used to create scatter plots. It takes in two sets of data points (x and y) and plots them as individual points on a Cartesian plane. The basic syntax for plt.scatter() is as follows:

```python
plt.scatter(x, y, s=None, c=None, marker=None, **kwargs)
```
- x: The data points for the x-axis.
- y: The data points for the y-axis.
- s: An optional parameter that specifies the size of the markers (data points).    
- c: An optional parameter that specifies the color of the markers.
- marker: An optional parameter that specifies the shape of the markers (e.g., 'o' for circles, 's' for squares).
- **kwargs: Additional keyword arguments that can be used to customize the plot (e.g., labels, title, grid).



````python
import numpy as np
x=np.random.randint(100,size=100)
print(x)

import numpy as np

#random generation of Dataset
x=np.random.randint(100,size=100)
y=np.random.randint(100,size=100)

# Creating a Scatter plot
plt.scatter(x,y,marker='o',s=30)

#Add labels and Title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')

#Show the plot
plt.show()
````

# Histograms
Histograms are used to represent the distribution of a dataset by dividing the data into bins and counting the number of data points that fall into each bin. They provide a visual representation of the frequency distribution of the data, allowing you to see patterns, skewness, and the overall shape of the distribution. Histograms are particularly useful for understanding the underlying distribution of a dataset and identifying any potential outliers or anomalies.  



### plt.hist()
hist() is a function in the Matplotlib library that is used to create histograms. It takes in a dataset and divides it into bins, counting the number of data points that fall into each bin to visualize the distribution of the data. The basic syntax for plt.hist() is as follows:

```python     
plt.hist(x, bins=None, range=None, density=False, **kwargs)
```
- x: The data points to be plotted in the histogram.
- bins: An optional parameter that specifies the number of bins or the edges of the bins.  
- range: An optional parameter that specifies the lower and upper range of the bins.
- density: An optional parameter that, if set to True, normalizes the histogram so that the area under the histogram equals 1.
- **kwargs: Additional keyword arguments that can be used to customize the plot (e.g., labels, title, grid).    



````python
x=np.random.normal(20,1,100)
print(x)



#Random normal distribution with mean=20 and Std=5
x=np.random.normal(20,5,1000)

#plotting the histogram
plt.hist(x,bins=13)

# Add title
plt.title('Histogram',fontsize=25)

# Display the plot
plt.show()

````

# Bar Chart 
Bar charts are used to represent categorical data with rectangular bars, where the length of each bar is proportional to the value it represents. They are useful for comparing different categories or groups and visualizing the distribution of categorical data. Bar charts can be displayed vertically or horizontally, and they can also be stacked or grouped to show multiple categories within the same chart.      


### plt.bar()
bar() is a function in the Matplotlib library that is used to create bar charts. It takes in categorical data and corresponding values to create rectangular bars that represent the data visually. The basic syntax for plt.bar() is as follows:

```python     
plt.bar(x, height, width=0.8, color=None, **kwargs)
```    
- x: The categorical data for the x-axis (e.g., categories or groups).
- height: The values corresponding to each category that determine the height of the bars.
- width: An optional parameter that specifies the width of the bars (default is 0.8).
- color: An optional parameter that specifies the color of the bars.  
- **kwargs: Additional keyword arguments that can be used to customize the plot (e.g., labels, title, grid).




````python
import matplotlib.pyplot as plt

# Sample data
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(100,size=5)

# Create a bar chart
plt.bar(categories, values, color='skyblue',width=0.7)

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart')

# Display the plot
plt.show()
````

# Pie Chart
Pie charts are used to represent the proportions of different categories in a dataset as slices of a circular pie. Each slice represents a category, and the size of each slice is proportional to the value it represents. Pie charts are useful for showing the relative proportions of different categories and visualizing how they contribute to the whole. However, they can be less effective than other types of charts (like bar charts) when there are many categories or when the differences between categories are small.       



### plt.pie()
pie() is a function in the Matplotlib library that is used to create pie charts. It takes in a dataset and corresponding labels to create a circular chart that represents the proportions of different categories visually. The basic syntax for plt.pie() is as follows:

```python     
plt.pie(x, labels=None, autopct=None, startangle=0, **kwargs)
```    
- x: The values corresponding to each category that determine the size of the slices.
- labels: An optional parameter that specifies the labels for each slice of the pie chart.
- autopct: An optional parameter that specifies a string or function to label the slices with their percentage values.
- startangle: An optional parameter that specifies the starting angle of the pie chart (default is 0 degrees).
- **kwargs: Additional keyword arguments that can be used to customize the plot (e.g., colors, title, grid).    








````python
import matplotlib.pyplot as plt

# Sample data
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(100,size=5)

# Create a Pie chart
plt.pie(values, labels=categories,explode=[0,0.1,0,0,0],autopct="%.1f%%",startangle=90,shadow=True)

# Add Title
plt.title('Bar Chart')

# Display the plot
plt.show()
````

# box Plot 
Box plots, also known as box-and-whisker plots, are used to display the distribution of a dataset based on five summary statistics: minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum. They provide a visual representation of the central tendency, variability, and skewness of the data, as well as any potential outliers. Box plots are particularly useful for comparing the distribution of multiple datasets or groups.       

### plt.boxplot()
boxplot() is a function in the Matplotlib library that is used to create box plots. It takes in a dataset and visualizes the distribution of the data based on its summary statistics. The basic syntax for plt.boxplot() is as follows:

```python     
plt.boxplot(x, notch=False, vert=True, patch_artist=False, **kwargs)
```    
- x: The data points to be plotted in the box plot.
- notch: An optional parameter that, if set to True, creates a notched box plot to indicate the confidence interval around the median.
- vert: An optional parameter that, if set to True, creates a vertical box plot (default is True).
- patch_artist: An optional parameter that, if set to True, fills the box with color (default is False).
- **kwargs: Additional keyword arguments that can be used to customize the plot (e.g., labels, title, grid).    




````python
import matplotlib.pyplot as plt

# Sample data
x=np.random.normal(200, 8, 1000)

#Five Number summary
IQR=np.quantile(x, 0.75) - np.quantile(x, 0.25)
print('Lower Fence : ', np.quantile(x, 0.25)-(1.5*IQR))
print('Q1 : ',np.quantile(x, 0.25))
print('Median : ',np.median(x))
print('Q3 : ',np.quantile(x, 0.75))
print('Upper Fence : ',np.quantile(x, 0.75)+(1.5*IQR))
print('\n')


# Create a Pie chart
plt.boxplot(x)

# Add Title
plt.title('BOX Plot')
plt.yticks([np.quantile(x, 0.25)-(1.5*IQR),np.quantile(x, 0.25),np.median(x),np.quantile(x, 0.75),np.quantile(x, 0.75)+(1.5*IQR)])

# Display the plot
plt.show()
````

# subplots
Subplots are used to create multiple plots within a single figure, allowing you to compare different datasets or visualize multiple aspects of the same dataset side by side. Subplots can be arranged in a grid format, and each subplot can contain a different type of plot (e.g., line plot, scatter plot, bar chart) to provide a comprehensive view of the data. Subplots are particularly useful for presenting complex data in a clear and organized manner.   

### plt.subplot()
subplot() is a function in the Matplotlib library that is used to create subplots within a single figure. It allows you to specify the number of rows and columns of subplots, as well as the index of the current subplot to be created. The basic syntax for plt.subplot() is as follows:

```python     
plt.subplot(nrows, ncols, index)
```    
- nrows: The number of rows of subplots in the figure.
- ncols: The number of columns of subplots in the figure.             
- index: The index of the current subplot to be created (starting from 1 and increasing in row-major order).    
- **kwargs: Additional keyword arguments that can be used to customize the plot (e.g., labels, title, grid).    




````python
#Creating the dataset
x=np.arange(100)

P1=np.sin(x)
P2=np.cos(x)

#1st subplot
plt.subplot(221)
plt.plot(P1)
plt.title('sine function')

#2nd Subplot
plt.subplot(222)
plt.plot(P2)
plt.title('cosine function')

# Display the subplot
plt.show()
````
