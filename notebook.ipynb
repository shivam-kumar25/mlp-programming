
## Keyboard Shortcuts in Notebooks


` Shift + Enter ` : Run the current cell and move to the next.

` Ctrl + Enter ` : Run the current cell and stay in the same cell.

` Alt + Enter ` : Run the current cell and insert a new cell below.
